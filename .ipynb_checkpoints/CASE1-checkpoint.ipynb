```python
df_m['dof'] = df_m['Date'].dt.dayofweek


def change_dname(num) :
    if num == 0:
        return 'Monday'
    elif num == 1:
        return 'Tuesday'
    elif num == 2:
        return 'Wednesday'
    elif num == 3:
        return 'Thursday'
    elif num == 4:
        return 'Friday'
    elif num == 5:
        return 'Saturday'
    elif num == 6:
        return 'Sunday'
    
df_m['Day'] = df_m['dof'].apply(change_dname)

df_m['Month'] = df_m['Date'].dt.month

df_m
```
